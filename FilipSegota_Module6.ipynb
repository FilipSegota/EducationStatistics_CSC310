```python
#importing libraries
import os
import glob
import pandas as pd
```

```python
#defining format of files to be combined
extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
```

```python
#combining files
combined_csv = pd.concat([pd.read_csv(f) for f in all_filenames])
combined_csv.to_csv("combined_csv.csv", index=False)
```

```python
#reading combined file
content = pd.read_csv('combined_csv.csv')
```

```python
#function for finding countries
def listed(obj):
    tmp = []
    
    for i in obj["Table Name"]:
        tmp.append(i)
    return tmp
```
